```python
# Q-23
new_df = df[['quantity','item_price']]
new_df
```

```python
# Q-24
df['new_price'] = df['item_price'].str[1:].astype('float')
df['new_price'].head()
```

```python
df.info()
```

```python
df['new_price'] = df['item_price'].str.replace('$', '').astype('float')
df[['item_price','new_price']].head()
```

```python
print(df['item_price'].dtype)
print(df['new_price'].dtype)
```

```python
# Q-25
len(df.loc[df['new_price']<=5])
```

```python
# Q-26
df_2 = df.loc[df['item_name']=='Chicken Salad Bowl'].reset_index(drop=True)
df_2.head(3)
```

```python
# Q-27
df.loc[(df['new_price']<=9) & (df['item_name']=='Chicken Salad Bowl')].head()
```

```python
# Q-28
df_3 = df.sort_values('new_price').reset_index(drop=True)
df_3.head(4)
```

```python
# Q-29
df.loc[df['item_name'].str.contains('Chips')].head()
```

```python
# Q-30
df.iloc[:,::2].head()
```

```python
# Q-33
df_4 = df.loc[(df.item_name=='Steak Salad') | (df.item_name=='Bowl')]
df_4.drop_duplicates('item_name')
```

```python
# Q-34
df_5 = df.loc[(df.item_name=='Steak Salad')|(df.item_name=='Bowl')]
df_5.drop_duplicates(subset='item_name', keep='last').head()
```
