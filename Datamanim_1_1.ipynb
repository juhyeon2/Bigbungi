```python
# 유튜브 인기동영상 데이터
import pandas as pd
url = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Q-1
df.groupby('channelTitle').size().sort_values(ascending=False)[:10]
```

```python
top10 = df.channelId.value_counts().head(10).index
ans = df.loc[df.channelId.isin(top10)].channelTitle.unique()
ans
```

```python
# Q-2 ???
def nolike(x):
  if x.likes > x.dislikes:
    return 'nolike'
  else:
    return 'like'

df['like_nolike'] = df.apply(nolike)
df.like_nolike.value_counts()
```

```python
###
ans = list(df[df.dislikes > df.likes].channelTitle.unique())
ans
```

```python
# Q-3 ###
change = df[['channelTitle','channelId']].drop_duplicates().channelId.value_counts()
target = change[change > 1]
len(target)
```

```python
# Q-4 ###
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
ans = df[df['trending_date2'].dt.day_name() == 'Sunday'].categoryId.value_counts()
ans.index[0]
```

```python
# Q-5 ##
ans = df.groupby(['categoryId',df['trending_date2'].dt.day_name()]).size().unstack()
ans.head(5)
```

```python
group = df.groupby([df['trending_date2'].dt.day_name(),'categoryId'], as_index=False).size()
answer = group.pivot(index='categoryId', columns='trending_date2')
answer.head(5)
```
